```python
import os
import spacy
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk.corpus import stopwords
import re

# Load Spacy model and download NLTK resources
nlp = spacy.load("en_core_web_sm")
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Set folder path and initialize stopwords
folder_path = "C:\\Users\\maheshwari\\Downloads\\comments1k\\comments1k"
stop_words_nltk = set(stopwords.words('english'))

# Define comment processing function
def process_comment(comment):
    sentences = sent_tokenize(comment)
    tokens = word_tokenize(comment)
    words = re.findall(r'\b\w+\b', comment)
    lemmatizer = WordNetLemmatizer()
    porter_stemmer = PorterStemmer()
    lemmatized_result = [lemmatizer.lemmatize(token) for token in tokens]
    stemmed_result = [porter_stemmer.stem(token) for token in tokens]

    return {
        'total_sentences': len(sentences),
        'total_tokens': len(tokens),
        'total_words': len(words),
        # lemmatized_result': lemmatized_result,
        'stemmed_result': stemmed_result,
        'lem_len':len(lemmatized_result),
        'stem_len':len(stemmed_result)
    }

# Process comments in each file
processed_comments = []
for filename in os.listdir(folder_path):
    if filename.endswith(".txt") and os.path.isfile(os.path.join(folder_path, filename)):
        with open(os.path.join(folder_path, filename), 'r', encoding='utf-8', errors='ignore') as file:
            comments = file.read()
            comment=comments.lower()
            if comment:
                processed_comments.append({'filename': filename, 'comment_info': process_comment(comment)})

# Calculate averages
total_comments = len(processed_comments)
averages = {
    'sentences_per_comment': sum(c['comment_info']['total_sentences'] for c in processed_comments) / total_comments,
    'tokens_per_comment': sum(c['comment_info']['total_tokens'] for c in processed_comments) / total_comments,
    'words_per_comment': sum(c['comment_info']['total_words'] for c in processed_comments) / total_comments
}

# Print results
print(f"Total comments: {total_comments}")
print(f"Avg sentences per comment: {averages['sentences_per_comment']}")
print(f"Avg tokens per comment: {averages['tokens_per_comment']}")
print(f"Avg words per comment: {averages['words_per_comment']}")

# Print the first processed comment for reference
if processed_comments:
    print("\nFirst processed comment:")
    print(processed_comments[0])
    print(processed_comments[1])
    print(processed_comments[2])
 
    
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.metrics import cohen_kappa_score
from gensim.corpora import Dictionary
from gensim.models import CoherenceModel
import os
def process_comment():
    lemmatized_output = [comment for comment_list in processed_comments for comment in comment_list['comment_info']]
    return lemmatized_output
# Function for LDA Topic Modeling
def lda_topic_modeling(documents, num_topics):
    print(documents)

    # Create a CountVectorizer to convert text data to a bag-of-words representation
    vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
    X = vectorizer.fit_transform(documents)
    # Convert the CountVectorizer output to a gensim Dictionary
    # gensim_dictionary = Dictionary([doc.split() for doc in document_strings])


    # Apply LDA
    lda = LatentDirichletAllocation(n_components=num_topics, random_state=42)
    lda.fit(X)

    return vectorizer, lda

# Function to print top words for each topic
def print_top_words(lda, vectorizer, n_words=10):
    feature_names = vectorizer.get_feature_names_out()
    for topic_idx, topic in enumerate(lda.components_):
        top_words_idx = topic.argsort()[:-n_words - 1:-1]
        top_words = [feature_names[i] for i in top_words_idx]
        print(f"Topic #{topic_idx + 1}: {', '.join(top_words)}")

# Function to get the most relevant topic for a document
def get_document_topic(lda, vectorizer, document):
    X = vectorizer.transform([document])
    topic_distribution = lda.transform(X)
    most_relevant_topic = topic_distribution.argmax()
    return most_relevant_topic

preprocessed_comments = lemmatized_output

# Question 1: Use LDA with different numbers of topics
num_topics_list = [15]
for num_topics in num_topics_list:
    print(f"\nLDA Topic Modeling with {num_topics} Topics:")
    vectorizer, lda = lda_topic_modeling(preprocessed_comments, num_topics)
    print_top_words(lda, vectorizer)


# # Question 2: Apply necessary text preprocessing techniques and evaluate topic modeling results

# # Question 3: Get the most relevant topic for specific documents
# document_filenames = ["0_9.txt", "1_7.txt", "2_9.txt"]
# for filename in document_filenames:  
#     document_path = os.path.join(folder_path, filename)
#     with open(document_path, 'r', encoding='utf-8', errors='ignore') as file:
#         document_content = file.read()
#         preprocessed_document = preprocess_text(document_content)  # Assuming preprocess_text is defined
#         most_relevant_topic = get_document_topic(lda, vectorizer, preprocessed_document)
#         print(f"\nMost relevant topic for {filename}: Topic #{most_relevant_topic + 1}")


```
